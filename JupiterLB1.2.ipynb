```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Telor/Desktop/shopping_trends.csv", encoding="cp1251")
#перевірка
print("Розмір таблиці:", df.shape)
print("\nПропуски:")
print(df.isnull().sum())
print("\nДублікати:", df.duplicated().sum())
df.columns = df.columns.str.lower().str.replace(" ", "_")
print("\nТипи стовпців:")
print(df.dtypes)
#таблиця
df.head(10)
```

```python
#Гістограма віку (age):
sns.histplot(df['age'], bins=20)
plt.title('Гістограма віку')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

```python
#Порівняння покупок
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='item_purchased', order=df['item_purchased'].value_counts().index, hue="gender", palette='pastel')
plt.title('Розподіл покупок')
plt.xlabel('Кількість покупок')
plt.ylabel('Товар')
plt.show()
```

```python
#Залежність між віком та сумою покупок 
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупки (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['review_rating'], bins=5, kde=True, color='green')
plt.title('Графік розподілу оцінок товарів')
plt.xlabel('Оцінка')
plt.ylabel('Кількість товарів')
plt.show()
```

```python
#Графік розподілу оцінок товарів (review_rating)
plt.figure(figsize=(8,5))
sns.histplot(df['review_rating'], bins=5, discrete=True)
plt.title('Графік розподілу оцінок товарів')
plt.xlabel('Оцінка')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='review_rating', y='purchase_amount_(usd)', color='red')
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Рейтинг')
plt.ylabel('Сума покупок')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='season', data=df, order=df['season'].value_counts().index)
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='size', order=df['size'].value_counts().index, hue='category', palette='muted')
plt.title('Розподіл розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='payment_method', data=df, order=df['payment_method'].value_counts().index)
plt.title('Кількість покупок відносно типу оплати')
plt.xlabel('Метод оплати')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', hue="season", palette="Set2")
plt.title('Покупки в різних категоріях за сезонами')
plt.xlabel('Категорія')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='previous_purchases', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між кількістю попередніх покупок і сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума покупки (USD)')
plt.show()
```

```python
import numpy as np

# Графік поверхні що вказує на залажність змінних
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

plt.figure(figsize=(10, 6))
ax = plt.axes(projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.colorbar(surf, ax=ax, shrink=0.5, aspect=5) 
plt.show()
```
